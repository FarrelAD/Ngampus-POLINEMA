Berikut adalah beberapa cara untuk menentukan fungsi turunan dari berbagai metode
## Metode Selisih Maju ‚è©
$$ f' = \frac{f(x + h) - f(x)}{h} $$
## Metode Selisih Mundur üîô
$$ f' = \frac{f(x) - f(x - h)}{h} $$
## Metode Selisih Tengahan üçî
$$ f' = \frac{f(x + h) - f(x - h)}{2h} $$
## Penentuan Diferensial Tingkat Tinggi üîù
### Turunan Kedua Selisih Tengahan
$$ f'' = \frac{f(x - h) -2f(x) + f(x + h)}{h^{2}} $$

# Implementasi berbagai metode ke dalam kode Python

```python
# Implementasi metode selisih mundur
def metode_selisih_mundur(fx, f_diff, f_approx_diff, h, x1, x2):
    x_values = []
    f_values = []
    f_approx_values = []
    errors = []
    
    iterasi = 1
    while x1 <= x2:
        x_values.append(x1)
        f_values.append(fx(x1))
        f_approx_values.append(f_approx_diff(x1, h))
        errors.append(hitung_error(f_diff(x1), f_approx_diff(x1, h)))
        
        print(f"{'Iterasi':<8}{'x':<25}{'f(x)':<25}{'f(x-h)':<25}{'f_approx_diff()':<25}{'f_diff()':<25}{'error':<10}")
        print(f"{iterasi:<8}{x1:<25}{fx(x1):<25}{fx(x1 - h):<25}{f_approx_diff(x1, h):<25}{f_diff(x1):<25}{hitung_error(f_diff(x1), f_approx_diff(x1, h)):<10}")

        x1 += h
        iterasi += 1

    return x_values, f_values, f_approx_values, errors


# Implementasi metode selisih maju
def metode_selisih_maju(fx, f_diff, f_approx_diff, h, x1, x2):
    x_values = []
    f_values = []
    f_approx_values = []
    errors = []
    
    iterasi = 1
    while x1 <= x2:
        x_values.append(x1)
        f_values.append(fx(x1))
        f_approx_values.append(f_approx_diff(x1, h))
        errors.append(hitung_error(f_diff(x1), f_approx_diff(x1, h)))
        
        print(f"{'Iterasi':<8}{'x':<25}{'f(x)':<25}{'f(x+h)':<25}{'f_approx_diff()':<25}{'f_diff()':<25}{'error':<10}")
        print(f"{iterasi:<8}{x1:<25}{fx(x1):<25}{fx(x1 + h):<25}{f_approx_diff(x1, h):<25}{f_diff(x1):<25}{hitung_error(f_diff(x1), f_approx_diff(x1, h)):<10}")

        x1 += h
        iterasi += 1
        
    return x_values, f_values, f_approx_values, errors

def metode_selisih_tengahan(fx, f_diff, f_approx_diff, h, x1, x2):
    x_values = []
    f_values = []
    f_approx_values = []
    errors = []
    
    iterasi = 1
    while x1 <= x2:
        x_values.append(x1)
        f_values.append(fx(x1))
        f_approx_values.append(f_approx_diff(x1, h))
        errors.append(hitung_error(f_diff(x1), f_approx_diff(x1, h)))
        
        print(f"{'Iterasi':<8}{'x':<25}{'f(x)':<25}{'f(x+h)':<25}{'f_approx_diff()':<25}{'f_diff()':<25}{'error':<10}")
        print(f"{iterasi:<8}{x1:<25}{fx(x1):<25}{fx(x1 + h):<25}{f_approx_diff(x1, h):<25}{f_diff(x1):<25}{hitung_error(f_diff(x1), f_approx_diff(x1, h)):<10}")

        x1 += h
        iterasi += 1

    return x_values, f_values, f_approx_values, errors

def metode_differensiasi_tinggi(fx, f_2nd_diff, f_2nd_approx_diff, h, x1, x2):
    x_values = []
    f_values = []
    f_2nd_diffs_values = []
    f_2nd_approx_diffs_values = []
    errors = []
    
    iterasi = 1
    while x1 <= x2:
        x_values.append(x1)
        f_values.append(fx(x1))
        f_2nd_diffs_values.append(f_2nd_diff(x1))
        f_2nd_approx_diffs_values.append(f_2nd_approx_diff(x1, h))
        errors.append(hitung_error(f_2nd_diff(x1), f_2nd_approx_diff(x1, h)))
        
        print(f"{'Iterasi':<8}{'x':<25}{'f(x)':<25}{'f_2nd_diff(x)':<25}{'f_2nd_approx_diff()':<25}{'error':<10}")
        print(f"{iterasi:<8}{x1:<25}{fx(x1):<25}{f_2nd_diff(x1):<25}{f_2nd_approx_diff(x1, h):<25}{hitung_error(f_2nd_diff(x1), f_2nd_approx_diff(x1, h)):<10}")

        x1 += h
        iterasi += 1

    return x_values, f_values, f_2nd_diffs_values, f_2nd_approx_diffs_values, errors

# Perhitungan error absolut
def hitung_error(f_diff, f_approx_diff):
    return abs(f_diff - f_approx_diff)
```

# Kode untuk visualisasi data

```python
########################## 
def visualize_fixed_graph(
    fx, x_range=(-5, 5), num_points=500, y_limits=(-10, 10), 
    x_limits=(-5, 5), vertical_lines=None, 
    title='Graph', x_label='x', y_label='y', 
    line_color='blue', equation_label='y = equation'
):
    x_values = np.linspace(x_range[0], x_range[1], num_points)
    y_values = fx(x_values)

    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label=equation_label, color=line_color)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.ylim(y_limits)
    plt.xlim(x_limits)
    
    if vertical_lines:
        for x_val, color, label in vertical_lines:
            plt.axvline(x=x_val, color=color, linestyle='--', label=label)
    
    plt.axhline(0, color='black', lw=0.5, ls='-')
    plt.axvline(0, color='black', lw=0.5, ls='-')
    
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()


##########################
def visualize_animate_data(x_values, f_values, f_approx_values, f_2nd_approx_values=None, interval=50):
    """Visualisasi data menggunakan matplotlib animation
    
    Parameters:
        x_values (list): List seluruh nilai x.
        f_values (list): List seluruh nilai f(x).
        f_approx_values (list): List seluruh nilai dari fungsi differensial.
        f_approx_second_values (list, optional): List seluruh nilai dari fungsi differensial kedua.
        interval (int): Interval untuk tiap frame. Secara default nilai interval = 50.
    """

    fig, ax = plt.subplots()
    line1, = ax.plot([], [], label='f(x)', color='blue')
    line2, = ax.plot([], [], label='f\'(x)', color='orange')

    ax.set_xlim(min(x_values), max(x_values))
    ax.set_ylim(min(min(f_values), min(f_approx_values)) - 1, max(max(f_values), max(f_approx_values)) + 1)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend()
    ax.grid()

    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)

    if f_2nd_approx_values is not None:
        line3, = ax.plot([], [], label='f\'\'(x)', color='green')
        ax.legend()

    def init():
        line1.set_data([], [])
        line2.set_data([], [])
        if f_2nd_approx_values is not None:
            line3.set_data([], [])
        return line1, line2

    def update(frame):
        line1.set_data(x_values[:frame + 1], f_values[:frame + 1])
        line2.set_data(x_values[:frame + 1], f_approx_values[:frame + 1])
        if f_2nd_approx_values is not None:
            line3.set_data(x_values[:frame + 1], f_2nd_approx_values[:frame + 1])
        return line1, line2

    anim = FuncAnimation(fig, update, frames=len(x_values), init_func=init, blit=True, repeat=True, interval=interval)

    video = anim.to_html5_video()
    html = display.HTML(video)
    display.display(html)
    plt.close()
```

# Soal 1
$ f(x) = 3x^{3} $ dengan metode selisih mundur dari range $ x=[-1,1] $ dengan
$ h=0.01 $ dan bandingkan dengan $ h=0.1 $, beri komentarmu.

```python
def f1(x):
    return 3 * x**3

def f1_diff(x):
    return 9 * x**2

def f1_approx_diff(x, h):
    return (f1(x - h) - f1(x)) / h

h1_a = 0.01
h1_b = 0.1
x1 = [-1, 1]
```

```python
# Ekseksui kode untuk pemecahan soal nomor 1 dengan nilai h = 0.1
x_values_1, f_values_1, f_approx_values_1, errors_1 = metode_selisih_mundur(f1, f1_diff, f1_approx_diff, h1_b, x1[0], x1[1])
visualize_animate_data(x_values_1, f_values_1, f_approx_values_1, interval=150)
```

```python
# Ekseksui kode untuk pemecahan soal nomor 1 dengan nilai h = 0.01
x_values_1, f_values_1, f_approx_values_1, errors_1 = metode_selisih_mundur(f1, f1_diff, f1_approx_diff, h1_a, x1[0], x1[1])
visualize_animate_data(x_values_1, f_values_1, f_approx_values_1, interval=50)
```

## Visualiasi grafik

```python
visualize_fixed_graph(
    f1,
    vertical_lines=[
        (x1[0], 'red', f'x = {x1[0]}'), 
        (x1[1], 'green', f'x = {x1[1]}')
    ],
    title='Graph persamaan y = 3x^3',
    equation_label='y = 3x^3'
)
```

Jika dilihat pada grafik, nilai tepat $ y = 0 $ kurang lebih berada di posisi $ x = 0 $. Dengan demikian, hasil perolehan data yang saya lakukan di atas dengan menggunakan metode selisih maju sudah tepat. Hasilnya sudah mendekati nilai sebenarnya.
